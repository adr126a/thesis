```python
#Parte 2 - punto 4

# File location and type
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# Dataframe orders
df_orders = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load("/FileStore/tables/LUMSA/orders.csv")

display(df_orders)

# Dataframe departments
df_departments = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load("/FileStore/tables/LUMSA/departments.csv")

display(df_departments)

# Dataframe products
df_products = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load("/FileStore/tables/LUMSA/products.csv")

display(df_products)
```

```python
#Parte 2 - punto 5

# Create temporary tables

temp_table_name_orders = "orders_csv"
df_orders.createOrReplaceTempView(temp_table_name_orders)

temp_table_name_departments = "departments_csv"
df_departments.createOrReplaceTempView(temp_table_name_departments)

temp_table_name_products = "products_csv"
df_products.createOrReplaceTempView(temp_table_name_products)
```

```python
#Parte 2 - punto 6

# Count dataframe rows

print("Orders rows:", df_orders.count())
print("Departments rows:", df_departments.count())
print("Products rows:", df_products.count())
```

```python
#Parte 2 - punto 7

from pyspark.sql.functions import *
ORDERS_DAY = df_orders.withColumn('order_dow', when(df_orders.order_dow == 0, "domenica"). 
                                               when(df_orders.order_dow == 1,"lunedì").
                                               when(df_orders.order_dow == 2,"martedì").
                                               when(df_orders.order_dow == 3,"mercoledì").
                                               when(df_orders.order_dow == 4,"giovedì").
                                               when(df_orders.order_dow == 5,"venerdì").
                                               when(df_orders.order_dow == 6,"sabato"))
ORDERS_DAY = ORDERS_DAY.withColumnRenamed('order_dow', 'day_of_week')
ORDERS_DAY.display()

temp_table_name_orders = "orders_csv"
ORDERS_DAY.createOrReplaceTempView(temp_table_name_orders)

print("Il numero totale di ordini è", df_orders.count())
```

```python
#Parte 3 - Punto 9

ordersXhour = spark.sql("SELECT COUNT(order_id) AS total_orders, order_hour_of_day FROM orders_day GROUP BY order_hour_of_day ORDER BY order_hour_of_day ASC")
ordersXhour.display()

print("Le ore in cui si effettuano più ordini sono quelle diurne(8:00 - 20:00).")
print("Al contrario in quelle notturne si effettuano meno ordini.")
print("Il maggior numero di ordini avviene alle 10:00, mentre il minor numero si ha alle 3:00.")
```

```python
#Parte 3 - Punto 10

productsXdepartments = spark.sql("SELECT * FROM products_csv AS P LEFT OUTER JOIN departments_csv AS D ON P.department_id=D.department_id")
productsXdepartments.display()

temp_table_productsXdepartments = "productsXdepartments_csv"
productsXdepartments.createOrReplaceTempView(temp_table_productsXdepartments)
```

```python
#Parte 3 - Punto 11

NProductsXdepartments = spark.sql("SELECT COUNT(product_id) AS Numero_prodotti, department FROM productsXdepartments_csv GROUP BY department ORDER BY Numero_prodotti DESC")
NProductsXdepartments.limit(10).display()
```
